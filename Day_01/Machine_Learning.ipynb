# Machine Learning

Machine learning is a subfield of artificial intelligence (AI) that enables systems to automatically learn and improve from examples and experience (aka data), without being explicitly programmed.

## Supervised Learning

Machines often learn from sample data that has both an example input and an example output. For example, a data input-output sample pair may be input data about an individual’s credit history, and their corresponding credit risk (either specified by a human expert or based on historical outcomes). Given enough of these input-output samples, the machine learns how to construct a “model“ (i.e., a logic or algorithm) that represents the relationship between an individual's credit history and how 'risky' they might be considered from a credit perspective, based on the samples it has seen. 

This class of machine learning is called “supervised learning”, since along with the input data, the desired outcome is also given, and the model is “supervised” or guided by the outcome in learning the associated model parameters. 

There are 2 types of supervised learning approaches:
 
a) **Regression** 

Like classification, regression is also about inputs and corresponding outputs. But outputs for classification are typically discrete types (cat, dog), outputs for regression are a general number.


b) **Classification**

The goal of a classification problem is to determine what group a given input belongs to. A classic example is categorizing animal pictures into a cat group and a dog group. In other words, it’s not a 0 or 1, but a sliding scale of possibility. For example, given a radiological image, a model could predict how many more years the associated individual will be sick or healthy.

## Unsupervised Learning

Another form of machine learning is "unsupervised learning". In unsupervised learning, the machine learns from data for which the outcomes are not known. It’s given input samples, but no output samples.

There are 2 types of unsupervised learning approaches:

a) **Clustering** 
[Source](https://medium.com/machine-learning-in-practice/a-gentle-introduction-to-machine-learning-concepts-cfe710910eb)

Given only the set of documents, the objective is to automatically learn how to cluster them into types. The model may cluster the sports documents in one group and the history documents in another, but it was never told explicitly what a sports or history document looked like, as it was never shown sample output data. In fact, once clustering is complete, the model still won’t know what a sports or history document is. All the model “knows” is that inputs in Group A are similar to each other, as are the inputs in Group B. It’s for humans to look at the clusters and decide whether and how they make sense.

b) **Dimensionality Reduction** 
[Source](https://towardsdatascience.com/unsupervised-learning-dimensionality-reduction-ddb4d55e0757)


Dimensionality is the number of variables, characteristics or features present in the dataset. This dimensions are represented as columns, and the goal is to reduce the number of them.

In most cases, those columns are correlated and, therefore, there is some information that is redundant which increase the dataset’s noise. This redundant information impacts negatively in Machine Learning model’s training and performance and that is why using dimensionality reduction methods becomes of paramount importance. It is a very useful way to reduce model’s complexity and avoid overfitting.

There are two main categories of dimensionality reduction:

1.   Feature Selection → we select a subset of features of the original dataset.

2.   Feature Extraction → we derive information from the orginal set to build a new feature subspace.

##Semi-supervised Learning

Supervised and unsupervised learning each have their pros and cons. Supervised learning is typically demanding from the standpoint of acquiring enough input-output labeled data for learning, while unsupervised learning is relatively less demanding in that regard because it only needs input data but its outcomes can be unpredictable. 

A middle ground is semi-supervised learning. In semi-supervised learning, a model uses the larger  unlabeled data to gain a general sense of the data’s structure, and a much smaller subset of labeled data to learn how to group and organize the whole dataset. [Source](https://medium.com/machine-learning-in-practice/a-gentle-introduction-to-machine-learning-concepts-cfe710910eb)

## Reinforcement Learning

Reinforcement learning is like unsupervised learning in the sense that outputs are usually not given.

The central concept of reinforcement learning is based around an “agent” (a computer or robot) that is interacting with an “environment” (here defined as everything that is not the agent).

The agent performs actions on the environment (for instance, a robot takes a step forward). Then, the environments will then provide some sort of feedback to the agent, usually in a form called the “reward.” [Source](https://medium.com/machine-learning-in-practice/a-gentle-introduction-to-machine-learning-concepts-cfe710910eb)

While supervised learning is about instructive feedback, reinforcement learning is about **evaluative feedback**.

```python
from sklearn.datasets import load_iris
from sklearn import tree
```

Iris data sets consists of 3 different types of irises’ (Setosa, Versicolour, and Virginica) petal and sepal length, stored in a 150x4 numpy.ndarray

The rows represent the samples. The columns represent the attributes (also called features) of the samples: Sepal Length, Sepal Width, Petal Length and Petal Width.

```python
# Load in our dataset
iris_data = load_iris()
```

Let's visualize the data first

```python
import matplotlib.pyplot as plt
```

```python
# we only take the first two features.
X = iris_data.data[:, :2]

# we get all the target for each sample
y = iris_data.target       

# we get minimum and maximum values so that we know what should be plot axis ticks 
# (this is only for visualization purpose)
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
  
# we get minimum and maximum values so that we know what should be plot axis ticks 
# (this is only for visualization purpose)
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5  

# we will be creating 2 figures, each of size 8 x 6 dpi
plt.figure(2, figsize=(8, 6))   
# clear any previous figures (if any), and make this the current figure
plt.clf()                       

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.title('Red = Setosa, Yellow = Versicolour, Gray = Virginica')

plt.show()
```

# Decision Trees

Decision trees for both classification and regression are super easy to use in Scikit Learn with a built in class.

We’ll initialise our decision tree for classification, which is also a built in class.

```python
# Initialize our decision tree object
classification_tree = tree.DecisionTreeClassifier()
```

Running training is then a simple one-liner! The $.fit(X, Y)$ function trains the model where X is the NumPy array of inputs and Y is the corresponding NumPy array of outputs.

```python
# Train our decision tree (tree induction + pruning)
classification_tree = classification_tree.fit(iris_data.data, iris_data.target)
```

# Graphviz

You have been acquainted with **Matplotlib**, now let's use a different visualization library in Python: **Graphviz**

```python
import graphviz 
```

Graphviz comes with a few options that will help in visualising the decision nodes and splits that the model learned which is super useful for understanding how it all works. 

```python
dot_data = tree.export_graphviz(classification_tree, out_file=None, feature_names=iris_data.feature_names, class_names=iris_data.target_names, filled=True, rounded=True, special_characters=True)  
```

Render the decision tree in a .pdf file

```python
graph = graphviz.Source(dot_data)  
graph.render("iris") 
```

### Answer 3

Gini:

Gini Index or Gini impurity measures the degree or probability of a particular variable being wrongly classified when it is randomly chosen.

If all the elements belong to a single class, then it can be called pure. The degree of Gini Index varies between 0 and 1,

'0' denotes that all elements belong to a certain class or there exists only one class (pure), and
'1' denotes that the elements are randomly distributed across various classes (impure).

A Gini Index of '0.5 'denotes equally distributed elements into some classes.

Approaches:

Random forest uses gini importance or mean decrease in impurity to calculate the importance of each feature.

## Question 6

What are the main similiarities and differences betweem Random Forest and Decision Trees? (Observe, research and/or discuss with team.) 


### Answer 6

##Similarities:

Random Forest and Decision Trees are both algorithms used for classification and regression problems. They use the concept of trees where the feature values and the corresponding target variable were used to divide the data into smaller groupings.

Both algorithms can handle non-linear relationships and missing values in the data.

Since the outcomes of both algorithms can be seen as a tree structure, they are both simple to comprehend and interpret.

##Differences:

Random Forest and Decision Trees have contrasting predictions. Random Forest mixes numerous decision trees to generate a final forecast, whereas Decision Trees only make one prediction based on the majority vote of the nodes.

Random Forest performs better than Decision Trees for large datasets.

Random Forest is less susceptible to overfitting since it reduces the impact of each decision tree's contribution to the final forecast. Decision trees can become complex and include data noise, making them susceptible to overfitting.

--Overfitting gives highley accurate output on training data that it fails to predict any other outcome in untrained data with accuracy
